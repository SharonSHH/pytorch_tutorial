# Matplotlib

Matplotlib helps us building graphs and visualisation figures to represent data.

### Python Matplotlib Terminology
The __Figure__ is complete window or the page the graph is drawn upon.
The __Axes__ is the area on which data is plotted. This can be X-Axis or Y-Axis etc.
The __Spines__ are the lines which connects Axes points.

### Types:
1. Line plot-straight line; curves: plt.plot()
2. Scatter plot: plt.scatter()
3. Histograms: plt.hist(a, b)
    a: list of values to plot
    b: Number of ranges to distribute these points
4. Bar: plt.bar()
5. Pie: plt.pie() 

```python
import matplotlib.pyplot as plt

year = [1950, 1975, 2000, 2018]
population = [2.12, 3.681, 5.312, 6.981]

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population')
plt.plot(year, population)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cos, sin = np.cos(X), np.sin(X)

plt.plot(X, cos, color='blue', label="cosine")
plt.plot(X, sin, color='red', label="sine")
plt.legend(loc='upper left', frameon=False)

plt.show()
```

```python
import matplotlib.pyplot as plt

values = [0, 1.2, 1.3, 1.9, 4.3, 2.5, 2.7, 4.3, 1.3, 3.9]
plt.hist(values, bins = 4)
plt.show()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
names = ('Tom', 'Dick', 'Harry', 'Jill', 'Meredith', 'George')
y_pos = np.arange(len(names))
speed = [8, 7, 12, 4, 3, 2]
 
plt.bar(y_pos, speed, align='center', alpha=0.5)
plt.xticks(y_pos, names)
plt.ylabel('Speed')
plt.title('Person')
 
plt.show()
```

```python
import matplotlib.pyplot as plt
 
# Data to plot
names = 'Tom', 'Dick', 'Harry', 'Jill', 'Meredith', 'George'
speed = [8, 7, 12, 4, 3, 2]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'blue']
explode = (0.1, 0, 0, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(speed, explode=explode, labels=names, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
